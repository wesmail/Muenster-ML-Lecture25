# Drawing Graphs with NetworkX

In this notebook, we will use **networkx** to create and visualize different types of graphs, including:

- Undirected graphs  
- Directed graphs (digraphs)  
- Weighted graphs  
- Graphs with self-loops  
- Directed Acyclic Graphs (DAGs)  
- ...  

**networkx** is a powerful and flexible Python library for handling complex network structures. It is especially handy for drawing and analyzing graphs of moderate size.

````python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])

plt.axis('off')
nx.draw_networkx(G,
                 pos=nx.spring_layout(G, seed=0),
                 node_size=600,
                 cmap='coolwarm',
                 font_size=14,
                 font_color='white',
                 node_color="dodgerblue"
                 )
````

````python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph object
G = nx.Graph()

# Add edges (as in the image)
edges = [
    ("A", "B"),
    ("A", "C"),
    ("B", "D"),
    ("B", "E"),
    ("C", "F"),
    ("C", "G")
]
G.add_edges_from(edges)

# Add self-loops
loops = ["A", "C", "E"]  # Example nodes with loops
G.add_edges_from([(node, node) for node in loops])

# Draw the graph
plt.axis('off')
pos = nx.spring_layout(G, seed=0)  # Seed for reproducible layout
nx.draw(G, pos, with_labels=True, node_size=600, node_color="dodgerblue", font_color="white", cmap='coolwarm', font_size=14)
````

````python
WG = nx.Graph()
WG.add_edges_from([('A', 'B', {"weight": 0.1}), ('A', 'C', {"weight": 1.2}), ('B', 'D', {"weight": 0.84}), ('B', 'E', {"weight": 0.0}), ('C', 'F', {"weight": -1.0}), ('C', 'G', {"weight": 4})])
labels = nx.get_edge_attributes(WG, "weight")

plt.axis('off')
nx.draw_networkx(WG,
                 pos=nx.spring_layout(G, seed=0),
                 node_size=600,
                 cmap='coolwarm',
                 font_size=14,
                 font_color='white',
                 node_color="dodgerblue"
                 )
nx.draw_networkx_edge_labels(WG, pos=nx.spring_layout(G, seed=0), edge_labels=labels)
````

````python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_node("x")
G.add_node("x^2")
G.add_node("x^2 + 5")

# Add operations (as edges with labels)
G.add_edge("x", "x^2", op="Power")
G.add_edge("x^2", "x^2 + 5", op="Addition")

# Define layout
pos = nx.spring_layout(G, seed=42)

# Draw nodes
nx.draw(G, pos, with_labels=True, node_color="dodgerblue", cmap='coolwarm', node_size=4000, font_weight="bold", font_color="white")

# Draw edges with operation labels
edge_labels = nx.get_edge_attributes(G, 'op')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="red", font_size=20)
nx.draw_networkx_edges(G, pos, arrows=True)

plt.axis("off")
````
